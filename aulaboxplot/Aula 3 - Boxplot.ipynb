```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


# Criando o array
Xi = np.array([1, 10, 3, 15, 2, 32, 43, 5, 63, 8, 11])

# Calculando a mdia
media = np.mean(Xi)

# Calculando a mediana
mediana = np.median(Xi)

# Calculando a moda contnua (usando KDE - Kernel Density Estimation)
kernel = stats.gaussian_kde(Xi)
x_range = np.linspace(Xi.min(), Xi.max(), 100)
densidade = kernel(x_range)
moda_continua = x_range[np.argmax(densidade)]

# Plotando o grfico da densidade com as medidas
plt.plot(x_range, densidade)
plt.axvline(media, color='red', linestyle='--', label='Mdia')
plt.axvline(mediana, color='green', linestyle='--', label='Mediana')
plt.axvline(moda_continua, color='purple', linestyle='--', label='Moda Contnua')
plt.xlabel('Valores')
plt.ylabel('Densidade')
plt.title('Densidade do Conjunto com Mdia, Mediana e Moda Contnua')
plt.legend()

# Plotando os valores de Xi prximos ao eixo x
for x in Xi:
  plt.plot(x, 0.01, 'bo')  # 'ko' plota pontos pretos ('k') como crculos ('o')

plt.show()

# Imprimindo os resultados
print("Mdia:", media)
print("Mediana:", mediana)
print("Moda Contnua:", moda_continua)
```
